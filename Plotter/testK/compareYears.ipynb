```python
import os
dir = "/scratch-cbe/users/alikaan.gueven/compare_2017_2018_MC/"

tFileDict_2017 = {
    "wjetstolnuht0100_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0100_2017_hist.root")),
    "wjetstolnuht0200_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0200_2017_hist.root")),
    # "wjetstolnuht0400_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0400_2017_hist.root")),
    "wjetstolnuht0600_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0600_2017_hist.root")),
    "wjetstolnuht0800_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0800_2017_hist.root")),
    "wjetstolnuht1200_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht1200_2017_hist.root")),
    # "wjetstolnuht2500_2017" : ROOT.TFile(os.path.join(dir, "wjetstolnuht2500_2017_hist.root")),

    # "zjetstonunuht0100_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0100_2017_hist.root")),
    # "zjetstonunuht0200_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0200_2017_hist.root")),
    # "zjetstonunuht0400_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0400_2017_hist.root")),
    # "zjetstonunuht0600_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0600_2017_hist.root")),
    # # "zjetstonunuht0800_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0800_2017_hist.root")),
    # # "zjetstonunuht1200_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht1200_2017_hist.root")),
    # "zjetstonunuht2500_2017" : ROOT.TFile(os.path.join(dir, "zjetstonunuht2500_2017_hist.root")),
}

tFileDict_2018 = {
    "wjetstolnuht0100_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0100_2018_hist.root")),
    "wjetstolnuht0200_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0200_2018_hist.root")),
    # "wjetstolnuht0400_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0400_2018_hist.root")),
    "wjetstolnuht0600_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0600_2018_hist.root")),
    "wjetstolnuht0800_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht0800_2018_hist.root")),
    "wjetstolnuht1200_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht1200_2018_hist.root")),
    # "wjetstolnuht2500_2018" : ROOT.TFile(os.path.join(dir, "wjetstolnuht2500_2018_hist.root")),

    # "zjetstonunuht0100_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0100_2018_hist.root")),
    # "zjetstonunuht0200_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0200_2018_hist.root")),
    # "zjetstonunuht0400_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0400_2018_hist.root")),
    # "zjetstonunuht0600_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0600_2018_hist.root")),
    # # "zjetstonunuht0800_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht0800_2018_hist.root")),
    # # "zjetstonunuht1200_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht1200_2018_hist.root")),
    # "zjetstonunuht2500_2018" : ROOT.TFile(os.path.join(dir, "zjetstonunuht2500_2018_hist.root")),
}


import re
tag = 'wjetstolnuht0100'
tag_2017 = tag + "_2017"
tag_2018 = tag + "_2018"

hist_vars = {}

for selection in tFileDict_2017[tag_2017].GetListOfKeys():
    hist_vars[selection.GetName()] = []
    for variable in getattr(tFileDict_2017[tag_2017], selection.GetName()).GetListOfKeys():
        hist_vars[selection.GetName()].append(variable.GetName())

compare_zjets_dir = os.path.join(dir, tag)




file_2017 = tFileDict_2017[tag_2017]
file_2018 = tFileDict_2018[tag_2018]


for sel, vars in hist_vars.items():
    compare_zjets_sel_dir = os.path.join(compare_zjets_dir, sel)
    os.makedirs(compare_zjets_sel_dir, exist_ok=True)
    for var in vars:
        zjets_2017 = getattr(getattr(file_2017, sel), var).Clone()
        zjets_2018 = getattr(getattr(file_2018, sel), var).Clone()

        zjets_2017.SetName(tag_2017)
        zjets_2018.SetName(tag_2018)


        # if zjets_2017.Integral() != 0: zjets_2017.Scale(1/zjets_2017.Integral())
        # if zjets_2017.Integral() != 0: zjets_2018.Scale(1/zjets_2018.Integral())
        c1 = ROOT.TCanvas("","",1600,1200)


        pattern2d = r"(_vs_)"
        if not bool(re.search(pattern2d, var)):
            zjets_2017.Draw("HIST")
            zjets_2018.Draw("HIST SAME")
        else:
            continue
            

        zjets_2017.SetLineColor(ROOT.kBlue)
        zjets_2018.SetLineColor(ROOT.kRed)

        n = zjets_2017.GetNbinsX()
        zjets_2017.GetXaxis().SetRange(0, n+1) 
        zjets_2018.GetXaxis().SetRange(0, n+1) 
        
        ymax_2017 = zjets_2017.GetMaximum()
        ymax_2018 = zjets_2018.GetMaximum()
        ymin_2017 = zjets_2017.GetMinimum()
        ymin_2018 = zjets_2018.GetMinimum()

        zjets_2017.GetYaxis().SetRangeUser(max(min(ymin_2017, ymin_2018)/2, 1e-3), 2*max(ymax_2017, ymax_2018))
        zjets_2018.GetYaxis().SetRangeUser(max(min(ymin_2017, ymin_2018)/2, 1e-3), 2*max(ymax_2017, ymax_2018)) 
        
        ROOT.gPad.BuildLegend(0.68,0.78,0.88,0.88)
        ROOT.gPad.SetLogy()
        c1.SaveAs(f"{compare_zjets_sel_dir}/{var}.png")
```
